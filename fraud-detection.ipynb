```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Loading the dataset
```

```python
credit_card_data = pd.read_csv('creditcard.csv')
```

```python
credit_card_data.head()
```

```python
# last 5 rows
credit_card_data.tail()
```

```python
# dataset information
credit_card_data.info()
```

```python
# checking no. of missing values in each column
credit_card_data.isnull().sum()
```

```python
# Distribution of the legit transactions and fraudulent transactions
credit_card_data['Class'].value_counts()
```

```python
# This dataset is highly unbalanced
# 0----> legit transaction , 1-----> fraudulent transaction
```

```python
# separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# Statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

```python
# Under-sampling : Build a sample dataset containing similar distribution of normal and fraud transactions
```

```python
legit_sample = legit.sample(n = 492)
```

```python
# Concatinating legit_sample and fraud data frames , here axis = 0  represents row wise
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
# splitting data into features and targets
```

```python
X = new_dataset.drop(columns = 'Class', axis = 1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
# train-test splitting of the data
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Model training
```

```python
model = LogisticRegression(max_iter = 1000)
```

```python
# training of the logistic regression with training data
model.fit(X_train, Y_train)
```

```python
# Model Evaluation
```

```python
# Accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy of the training data :', training_data_accuracy)
```

```python
# Accuracy on test data
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy of the test data :', test_data_accuracy)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
y_pred = model.predict(X_test)
```

```python
print(confusion_matrix(Y_test, y_pred))
```

```python
print(classification_report(Y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
print("Accuracy:", accuracy_score(Y_test, y_pred))
print("Precision:", precision_score(Y_test, y_pred))
print("Recall:", recall_score(Y_test, y_pred))
print("F1 Score:", f1_score(Y_test, y_pred))
```
